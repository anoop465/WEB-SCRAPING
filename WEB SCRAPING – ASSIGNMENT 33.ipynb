5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

def scrape_coordinates(city_name):
    # Initialize Chrome WebDriver
    driver = webdriver.Chrome()

    # Open Google Maps
    driver.get("https://www.google.com/maps")

    # Wait for the search box to load
    search_box = driver.find_element_by_css_selector("input[name='q']")
    search_box.clear()

    # Search for the city
    search_box.send_keys(city_name)
    search_box.send_keys(Keys.RETURN)

    # Wait for the map to load
    time.sleep(5)

    # Extract latitude and longitude
    try:
        # Find the coordinates element
        coordinates_element = driver.find_element_by_css_selector("meta[itemprop='geo.position']")
        content = coordinates_element.get_attribute("content")
        latitude, longitude = content.split(",")
        latitude = float(latitude)
        longitude = float(longitude)
        print(f"Coordinates for {city_name}: Latitude {latitude}, Longitude {longitude}")
    except Exception as e:
        print(f"Failed to retrieve coordinates for {city_name}: {e}")

    # Close the browser
    driver.quit()

if __name__ == "__main__":
    city_name = input("Enter the name of the city to search on Google Maps: ")
    scrape_coordinates(city_name)
```
